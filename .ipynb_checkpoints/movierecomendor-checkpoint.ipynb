```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_style('dark')
%matplotlib inline
```

```python
data_rating= pd.read_csv("ratings.csv")
```

```python
data_rating
```

```python
data_movies=pd.read_csv("movies.csv")
```

```python
data_movies
```

```python
data_m_rating=pd.merge(data_rating,data_movies,on='movieId')
```

```python
data_m_rating
```

```python
data_m_rating.groupby('title')['rating'].mean().head()
```

```python
data_m_rating.groupby('title')['rating'].mean().sort_values(ascending=False).head(100)
```

```python
rating_mean_count=pd.DataFrame(data_m_rating.groupby('title')['rating'].count().sort_values(ascending=True))
```

```python
rating_mean_count['rating_counts']=pd.DataFrame(data_m_rating.groupby('title')['rating'].mean())
```

```python
rating_mean_count
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['patch.force_edgecolor']=True
rating_mean_count['rating'].hist(bins=50)
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['patch.force_edgecolor']=True
rating_mean_count['rating_counts'].hist(bins=50)
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['patch.force_edgecolor']=True
sns.jointplot(x='rating_counts',y='rating',data=rating_mean_count, alpha=0.4)
```

```python
movie_rating_user =data_m_rating.pivot_table(index='userId', columns='title',values='rating')
```

movie_rating_user

```python
movie_rating_user
```

```python
Forrest_Gump_ratings=movie_rating_user['Forrest Gump (1994)']
```

```python
Forrest_Gump_ratings
```

```python
movies_similar_forrest_gump= movie_rating_user.corrwith(Forrest_Gump_ratings)
pul_gump_corr= pd.DataFrame(movies_similar_forrest_gump, columns=['correlation'])
pul_gump_corr.dropna(inplace=True)
pul_gump_corr
```

```python
pul_gump_corr.sort_values('correlation', ascending=False)
```

```python
pul_gump_corr = pul_gump_corr.join(rating_mean_count['rating_counts'])
pul_gump_corr.head(100)
```

```python
pul_gump_corr[pul_gump_corr['rating_counts']>100].sort_values('correlation',ascending=False).head()
```
